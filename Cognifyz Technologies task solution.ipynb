```python
# Visualize the locations of restaurants on a map using latitude and longitude information.

import folium
```

```python
# Initialize the map centered around the average latitude and longitude of all restaurants
# You can adjust the starting location based on your dataset
map_center = [data['Latitude'].mean(), data['Longitude'].mean()]
restaurant_map = folium.Map(location=map_center, zoom_start=12)

# Add markers for each restaurant location
for index, row in data.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=row['Restaurant Name']  # Replace 'RestaurantName' with the actual column name for restaurant names
    ).add_to(restaurant_map)

# Save the map as a HTML file
restaurant_map.save('restaurant_map.html')
```

### **Task 1**

```python
# Determine the percentage of restaurants that offer table booking and online delivery.

# Filter restaurants that offer both table booking and online delivery
both_services = data[(data['Has Table booking'] == 'Yes') & (data['Has Online delivery'] == 'Yes')]

# Calculate the percentage of restaurants offering both services
percentage_both_services = (len(both_services) / len(data)) * 100

print(f"Percentage of Restaurants Offering Both Table Booking and Online Delivery: {percentage_both_services:.2f}%")
```

```python
# Compare the average ratings of restaurants with table booking and those without.


table_booking_yes = data[data['Has Table booking'] == 'Yes']
table_booking_no = data[data['Has Table booking'] == 'No']

# Calculate average ratings for restaurants with and without table booking
avg_rating_with_booking = table_booking_yes['Aggregate rating'].mean()
avg_rating_without_booking = table_booking_no['Aggregate rating'].mean()

print(f"Average Rating for Restaurants with Table Booking: {avg_rating_with_booking:.2f}")
print(f"Average Rating for Restaurants without Table Booking: {avg_rating_without_booking:.2f}")
```

```python
# Analyze the availability of online delivery among restaurants with different price ranges.

# Group by 'Price range' and calculate the percentage of restaurants with online delivery
online_delivery_by_price = data.groupby('Price range')['Has Online delivery'].value_counts(normalize=True).unstack() * 100

# Add percentage signs to the table
online_delivery_by_price = online_delivery_by_price.map(lambda x: f"{x:.2f}%")

# Display the results
print("Percentage of Restaurants Offering Online Delivery by Price Range:")
print(online_delivery_by_price)
```

### **Task 2**

```python
# Determine the most common price range among all the restaurants.

from scipy import stats
# Find the most common price range
most_common_price_range = stats.mode(data['Price range'])

print(f"The most common price range is {most_common_price_range[0]} which appears {most_common_price_range[1]} times")
```

```python
# Calculate the average rating for each price range


avg_rating_by_price_range = data.groupby('Price range')['Aggregate rating'].mean()

print("Average Rating for Each Price Range:")
print(avg_rating_by_price_range)
```

```python
# Identify the color that represents the highest average rating among different price ranges.


# Find the price range with the highest average rating
highest_avg_rating_range = avg_rating_by_price_range.idxmax()

# Filter the rows for that price range
highest_avg_rating_rows = data[data['Price range'] == highest_avg_rating_range]

# Find the most common rating color for that price range
most_common_color = highest_avg_rating_rows['Rating color'].mode()[0]

print(f"The price range with the highest average rating is: {highest_avg_rating_range}")
print(f"The color representing the highest average rating is: {most_common_color}")
```

```python
# Extract additional features from the existing columns, such as the length of the restaurant name or address.


# Extract the length of the restaurant name
data['Name Length'] = data['Restaurant Name'].apply(len)

# Extract the length of the address
data['Address Length'] = data['Address'].apply(len)

print(data[['Restaurant Name', 'Name Length', 'Address', 'Address Length']].head())
```

```python
# Create new features like "Has Table Booking" or "Has Online Delivery" by encoding categorical variables.

# Encode binary columns as 1 for 'Yes' and 0 for 'No'
binary_columns = ['Has Table booking', 'Has Online delivery', 'Is delivering now', 'Switch to order menu']
for col in binary_columns:
    data[col] = data[col].map({'Yes': 1, 'No': 0})
```
